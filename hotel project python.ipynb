```python
# changing date type
df['booking_date']=pd.to_datetime(df['booking_date'])
df['check_in_date']=pd.to_datetime(df['check_in_date'])
df['check_out_date']=pd.to_datetime(df['check_out_date'])
```

```python
df.info()
```

```python
# finding new fields 
df['stay_length']=(df['check_out_date']-df['check_in_date']).dt.days
```

```python
# finding booking month 
df['month'] = df['booking_date'].dt.month
```

```python
# booking year 
df['year'] = df['booking_date'].dt.year
```

```python
# finding profit on each room 
df['profit'] = df['selling_price']-df['costprice']
```

```python
# Exploratory data analysis 
```

```python
# Booking status count 
sns.countplot(x='booking_status',data=df)
plt.show()
# observation = Confirmed ≈72% Cancelled ≈20% Failed ≈8%
```

```python
#Channel analysis
pd.crosstab(df['booking_channel'],
df['booking_status'],normalize='index')*100

#Observed:
#Web: Highest confirmed bookings, Lowest cancellation
#Travel Agent: Highest cancellation
#Mobile App: Moderate performance
```

```python
#Room type analysis
sns.countplot(x='room_type',hue='booking_status',data=df)
plt.xticks(rotation=45)
plt.show()

# Observation: Standard rooms have more cancellations. Deluxe and Suite perform better.
```

```python
# Monthly booking trend
monthly=df.groupby('month')['booking_value'].sum()
monthly.plot()
plt.show()

# observation = peak at holidays seasons, concentrate more on low holidays time 
```
